```python
import pandas as pd
df = pd.read_csv('C:/1.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.subplots(figsize=(15,10))
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

# Data where Depth 0

```python
data = df[df['Depth (m)']  == '0']
data
```

# break into 5 classes

```python
to_ten = data[data['DGM (pg/L)']  <= 10]
len(to_ten)
```

```python
ten_tw = data[(data['DGM (pg/L)']  > 10) & (data['DGM (pg/L)']  <= 20)]
len(ten_tw)
```

```python
tw_th = data[(data['DGM (pg/L)']  > 20) & (data['DGM (pg/L)']  <= 30)]
len(tw_th)
```

```python
th_fr = data[(data['DGM (pg/L)']  > 30) & (data['DGM (pg/L)']  <= 40)]
len(th_fr)
```

```python
fr = data[(data['DGM (pg/L)']  > 40)]
len(fr)
```

```python
Tto_ten = data[data['Total Hg (ng/L)']  <= 0.10]
len(Tto_ten)
```

```python
Tten_tw = data[(data['Total Hg (ng/L)']  > 0.10) & (data['Total Hg (ng/L)']  <= 0.20)]
len(Tten_tw)
```

```python
Ttw_th = data[(data['Total Hg (ng/L)']  > 0.20) & (data['Total Hg (ng/L)']  <= 0.30)]
len(Ttw_th)
```

```python
Tth_fr = data[(data['Total Hg (ng/L)']  > 0.30) & (data['Total Hg (ng/L)']  <= 0.40)]
len(Tth_fr)
```

```python
Tfr = data[(data['Total Hg (ng/L)']  > 0.40)]
len(Tfr)
```

```python
Mto_ten = data[data['MeHg (pg/L)']  <= 10]
len(Mto_ten)
```

```python
Mten_tw = data[(data['MeHg (pg/L)']  > 10) & (data['MeHg (pg/L)']  <= 20)]
len(Mten_tw)
```

```python
Mtw_th = data[(data['MeHg (pg/L)']  > 20) & (data['MeHg (pg/L)']  <= 30)]
len(Mtw_th)
```

```python
Mth_fr = data[(data['MeHg (pg/L)']  > 0.30) & (data['MeHg (pg/L)']  <= 0.40)]
len(Mth_fr)
```

```python
Mfr = data[(data['DGM (pg/L)']  > 40)]
len(Mfr)
```
